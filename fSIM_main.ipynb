```python
# Loading data

### simulation data
# f = open('res_speckle_shift.pckl', 'rb')
# pscrop,lambda_f,NA_obj,I_image,Ic_image,upsamp_factor,bg,pixel_shift_stack,speckle_intensity_shift_crop = pickle.load(f)
# f.close()

### experiment data
f = open('fluorescent_data.pckl', 'rb')
pscrop,lambda_f,NA_obj,I_image,Ic_image, upsamp_factor, bg = pickle.load(f)
f.close()

```

```python
# Upsampling image

I_image_up = image_upsampling(I_image, upsamp_factor, bg)
```

```python
display_image_movie(I_image_up, frame_num=40, size=(7,7), pause_time=0.0001)
```

```python
# Image registration to get shift
xshift, yshift = image_registration(Ic_image, 100, upsamp_factor)

plt.figure(figsize=(8,8))
plt.plot(xshift,yshift,'w')
plt.axis('equal')
plt.xlabel('xshift (px)')
plt.ylabel('yshift (px)')
```

```python
# Initialize the solver object

fSIM_sol = fSIM_solver(I_image_up, xshift, yshift, 80, lambda_f,\
                       pscrop, upsamp_factor, NA_obj, 0.5, 20)
```

```python
# Run the iterative algorithm
figsize = (18,6)
fSIM_sol.iterative_algorithm(I_image_up, 1, 1e3, 1, 1/12, figsize)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(np.maximum(0,fSIM_sol.I_obj), cmap='gray', clim=(0, np.max(fSIM_sol.I_obj)/2))
```
